# 1.Create various visualizations for deeper insights (e.g., box plots, violin plots, pair plots).

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("encoded_dataset.csv")
data
```

```python
# Drop irrelevant columns (e.g., customerID)
data.drop(columns=['customerID'], inplace=True)
```

```python
# Box plot for numerical features
plt.figure(figsize=(12, 8))
sns.boxplot(data=data.drop(columns=['Churn']), orient='h')
plt.title('Box Plot of Numerical Features')
plt.show()
```

```python
# Violin plot for numerical features
plt.figure(figsize=(12, 8))
sns.violinplot(data=data.drop(columns=['Churn']), orient='h')
plt.title('Violin Plot of Numerical Features')
plt.show()
```

```python
# Pair plot for selected features
selected_features = ['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']
sns.pairplot(data[selected_features], hue='Churn')
plt.suptitle('Pair Plot of Selected Features', y=1.02)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("encoded_dataset.csv")

# Drop irrelevant columns (e.g., customerID)
data.drop(columns=['customerID'], inplace=True)

# Create various visualizations

# Box plot for numerical features
plt.figure(figsize=(12, 8))
sns.boxplot(data=data.drop(columns=['Churn']), orient='h')
plt.title('Box Plot of Numerical Features')
plt.show()

# Violin plot for numerical features
plt.figure(figsize=(12, 8))
sns.violinplot(data=data.drop(columns=['Churn']), orient='h')
plt.title('Violin Plot of Numerical Features')
plt.show()

# Pair plot for selected features
selected_features = ['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']
sns.pairplot(data[selected_features], hue='Churn')
plt.suptitle('Pair Plot of Selected Features', y=1.02)
plt.show()
```
