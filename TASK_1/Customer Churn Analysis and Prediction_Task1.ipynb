# Importing Libraries

```python
import pandas as pd
import numpy as np
import sklearn
```

# Loading DataSet

```python
data=pd.read_csv("Customer_Churn_analysis.csv")
data
```

# 1.Handle missing values in the dataset.

```python
# Identify missing values
missing_values = data.isnull().sum()
missing_values
```

```python
# Save cleaned data
data.to_csv("Customer_Churn_analysis.csv", index=False)
```

# 2.Convert categorical variables into numerical representations using techniques like one-hot encoding.

```python
# Identify categorical variables
categorical_cols = ['gender','MultipleLines', 'InternetService','Contract','PaymentMethod']
```

```python
# Apply one-hot encoding
data_encoded = pd.get_dummies(data, columns=categorical_cols)
data_encoded 
```

```python
# Save encoded data
data_encoded.to_csv("encoded_dataset.csv", index=False)
```
